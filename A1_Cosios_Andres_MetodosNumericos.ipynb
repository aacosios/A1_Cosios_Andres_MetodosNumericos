## 2. Desarrolle en Python un programa para calcular la inversa de matrices de dimensión 2 x 2. No olvide colocar comentarios en su programa.

```python
#Imprtamos las librerias necesarias para realizar los ejercicios de la actividad previamente las instalamos en python
import matplotlib.pyplot as plt
import math
import numpy as np
from scipy import linalg
from math import pi
```

```python
# Declaramos la matriz A(2x2) y la imprimimos
A = np.array([[2, 3],[8,5]])
print(A)
```

```python
# Con la libreria linalg calculamos la matriz inversa y la imprimimos en B
B = linalg.inv(A)
print(B)
```

### Según la teoría podemos comprobar la matriz inversa de la siguiente manera:
### El producto de la matriz orginal por su inversa será igual a la matriz identidad
$$I=A . A^{-1} $$

```python
# Comprobamos el resultado calculando la matriz identidad segun la teoria indicada anteriormente
I = A.dot(B)
print(I)
```

## 4. Sea la función
$$ f(x)= \sqrt{x^2+2x+1} $$
## calcule, $f(2.045)$, con un error de 0.00005 (utilice una serie de Taylor). 

### La serie de taylor es igual a:

$$ f_{(x)}= f_{(x_0)}+f'_{(x_0)}(x-x_0)+f"_{(X_0)} \frac{(x-x_0)^2}{2!}+...+f_{(x_0)}^{(n)}\frac{(x-x_o)^n}{n!}$$

### Calculamos las derivadas:

$$ f(x)= \sqrt{x^2+2x+1} $$

$$ f'(x)= \frac {x+1}{\sqrt{x^2+2x+1}}$$

$$ f"(x)=\frac{1}{\sqrt{x^2+2x+1}}-\frac{(x+1)(2x+2)}{2(x^2+2x+1)^{\frac{3}{2}}}$$

$$ f'"(x)=0$$

### Se evalua en cada derivada con $x_0=1$,:

$$ f(x)\rightarrow f(1)=2 $$

$$ f'(x)\rightarrow f'(1)=1$$

$$ f"(x)\rightarrow f"(1)=0$$

$$ f"'(x)\rightarrow f"'(1)=0$$

### Reemplazando en la serie de Taylor:

$$f(2.045)\approx p(2.045)=2+2.045+0+0 $$

$$f(2.045)\approx p(2.045)=4.045 $$

## 6. Diseñe un código que encuentre el $\sin\left(\frac{\pi}{3}\right)$, a través del desarrollo de Taylor, truncar cuando $n=50$

```python
#asignamos los valores a las variables a utilizar
x=pi/3
n=int(50)
# inicializamos un acumulador en 0
senx=0.0
# creamos un ciclo for para acumular y sumar los terminos
for k in range(n):
    senx=senx + (-1)**k * x**(2*k+1)/math.factorial((2*k+1))
    # Imprimimos en pantalla los valores de k y el valor de cada termino 
    print(k,senx)
```
